```r
#bridge crossing times according to park officer records
bridge.times<-read.csv("https://raw.githubusercontent.com/norberello/gibbons-crossing-bridges/main/bridgingtimes.csv")
```

```r
head(bridge.times)
```

```r
library(ggplot2)
```

```r
ggplot(bridge.times,aes(Time))+
  geom_histogram(colour = "transparent", fill = "dark grey",
                 binwidth = 1,alpha=1)+theme_bw()+
  scale_x_continuous(breaks = c(4:18))+
  geom_density(aes(y=..count..*1),col="chocolate",size=1.5)+
  ylab("crossing frequency")+xlab("time(24h)")+
  facet_grid(~Bridge)+theme_bw()+
  theme(text = element_text(size = 18))
```
